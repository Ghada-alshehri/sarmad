```python
# albahtry
import os
import re
import markdown
from flask import Flask, request, jsonify, render_template_string, send_from_directory
from flask_ngrok import run_with_ngrok
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.schema import Document
import getpass
import ibm_watsonx_ai
from ibm_watsonx_ai.foundation_models import Model
from gtts import gTTS
import threading

# Setup Flask app
app = Flask(__name__)
run_with_ngrok(app)  # Start ngrok when app is run

# Directory for markdown files
directory = 'data/markdown_files'
os.makedirs(directory, exist_ok=True)

# Check if markdown files exist; if not, create them
if not any(fname.endswith('.md') for fname in os.listdir(directory)):
  #4928
    for i in range(0, 1175):
        title = df['poet_name'].iloc[i]
        content = df['poem_text'].iloc[i]

        if isinstance(content, str):
            lines = content.split('\n')
            markdown_content = f"# {title}\n\n" + "\n".join([line + "  " for line in lines]) + "\n\n"
            with open(f'{directory}/{i}.md', 'w', encoding='utf-8') as file:
                file.write(markdown_content)

# Read back the markdown files
markdown_texts = []
for filename in os.listdir(directory):
    if filename.endswith(".md"):
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
            markdown_content = file.read()
            html_content = markdown.markdown(markdown_content)
            markdown_texts.append(html_content)

# Function to chunk poem text
def chunk_poem_text(poem_text, lines_per_chunk=4, overlap=1):
    lines = poem_text.split('\n')
    chunks = []
    for i in range(0, len(lines), lines_per_chunk - overlap):
        chunk = lines[i:i + lines_per_chunk]
        if chunk:
            chunks.append('\n'.join(chunk))
    return chunks

# Create documents
documents = []
for markdown in markdown_texts:
    chunks = chunk_poem_text(markdown, lines_per_chunk=2, overlap=1)
    for chunk in chunks:
        documents.append(Document(page_content=chunk, metadata={}))

# Initialize the embedding model
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Check if Chroma database exists; if not, create it
chroma_db_dir = "./chroma_db"
if not os.path.exists(chroma_db_dir):
    db = Chroma(embedding_function=embedding_function, persist_directory=chroma_db_dir)


    # Add documents to the vector store
    #41666
    max_batch_size = 41666
    for i in range(0, len(documents), max_batch_size):
        batch = documents[i:i + max_batch_size]
        db.add_documents(batch)
    db.persist()
else:
    db = Chroma(embedding_function=embedding_function, persist_directory=chroma_db_dir)

def get_credentials():
    return {
        "url": "https://eu-de.ml.cloud.ibm.com",
        "apikey": getpass.getpass("Please enter your IBM WatsonX API key (hit enter): ")
    }

model_id = "sdaia/allam-1-13b-instruct"
parameters = {
    "decoding_method": "greedy",
    "max_new_tokens": 60,
    "repetition_penalty": 1
}
project_id = "9dcc614b-3e3e-4fce-b016-692b7c742644"
model = Model(
    model_id=model_id,
    params=parameters,
    credentials=get_credentials(),
    project_id=project_id,
)

def classify_question(question):
    """Simple classifier to detect if the question is poetry-related or about Abu Nawas."""
    poetry_keywords = ["poem", "verse", "write", "poetry", "شعر", "قصيدة"]
    abu_nawas_keywords = ["الطائي","البحتري"]

    if any(keyword in question.lower() for keyword in poetry_keywords):
        return "poetry"
    elif any(keyword in question for keyword in abu_nawas_keywords):
        return "abu_nawas"
    else:
        return "general"

def query_model(question):
    question_type = classify_question(question)

    if question_type == "poetry":
        similarity_search_results = db.similarity_search_with_score("البحتري", k=5)
        retrieved_poems = "\n\n".join([f"[Retrieved Poem {i+1}]\n{doc.page_content}" for i, (doc, _score) in enumerate(similarity_search_results)])
        # formatted_prompt = f" استنادًا إلى القصائد التالية، قم بكتابة قصيدة جديدة من اربع سطور : \n{retrieved_poems}\nالآن، اكتب قصيدة جديدة بأسلوبه."
        formatted_prompt = f"""
        استنادًا إلى القصائد التالية، قم بكتابة قصيدة جديدة من أربع سطور حيث:
        - السطر الأول ينتهي بحرف مماثل للسطر الثالث.
        - السطر الثاني ينتهي بحرف مماثل للسطر الرابع.
        \n{retrieved_poems}\n
        الآن، اكتب قصيدة جديدة بأسلوبه مع الالتزام بالقافية المذكورة.
        """
    elif question_type == "abu_nawas":
        formatted_prompt = f"أجب عن السؤال التالي بتفصيل: {question}"
    else:
        return ["أنا آسف، لا أستطيع مساعدتك بهذا."]

    generated_response = model.generate_text(prompt=formatted_prompt, guardrails=False)
    clean_response = re.sub(r'<.*?>', '', generated_response).strip()

    if question_type == "poetry":
        return clean_response.splitlines()
    else:
        return [clean_response]

def gtts_tts(text, lang='ar'):
    tts = gTTS(text=text, lang=lang, slow=False)
    audio_file_path = f"static/output_{threading.get_ident()}.mp3"
    tts.save(audio_file_path)
    return audio_file_path

@app.route('/')
def home():
    return render_template_string('''
        <!DOCTYPE html>
        <html lang="ar">
        <head>
            <meta charset="UTF-8">
            <style>
                body { font-family: Arial, sans-serif; background-color: #1a1a2e; color: white; }
                .chat-section {
                    background-color: #2a2a3c;
                    border-radius: 10px;
                    width: 90%;
                    max-width: 600px;
                    padding: 20px;
                    box-shadow: 0 0 15px rgba(0, 0, 0, 0.3);
                    margin: 0 auto;
                }
                .profile-section {
                    text-align: center;
                    margin-bottom: 20px;
                }
                .profile-section h1 {
                    font-size: 2em;
                    color: #00d4ff;
                    text-shadow: 0 0 5px #00d4ff;
                }
                .chat-messages {
                    max-height: 400px;
                    overflow-y: auto;
                    margin-bottom: 20px;
                }
                .chat-bubble {
                    background-color: #3a3a4f;
                    color: white;
                    padding: 10px 15px;
                    border-radius: 15px;
                    margin-bottom: 10px;
                    max-width: 70%;
                    display: flex;
                    align-items: center;
                    box-shadow: 0 4px 15px rgba(0, 0, 0, 0.2);
                }
                .chat-bubble.user {
                    background-color: #00d4ff;
                    margin-left: auto;
                    text-align: right;
                    flex-direction: row-reverse;
                    box-shadow: 0 4px 15px rgba(0, 212, 255, 0.3);
                }
                .chat-bubble.bot {
                    background-color: #3a3a4f;
                    margin-right: auto;
                    text-align: left;
                    flex-direction: row;
                }
                .chat-input-container {
                    display: flex;
                    margin-top: 20px;
                }
                .chat-input {
                    flex: 1;
                    padding: 12px;
                    border-radius: 5px;
                    border: none;
                    outline: none;
                    background-color: #3a3a4f;
                    color: white;
                }
                .send-button {
                    background-color: #00d4ff;
                    color: #1a1a2e;
                    border: none;
                    padding: 10px 20px;
                    border-radius: 5px;
                    cursor: pointer;
                    transition: background-color 0.3s;
                }
                .send-button:hover {
                    background-color: #00a6c9;
                }
                .mic-button {
                    background-color: #00c6ff;
                    border: none;
                    padding: 10px;
                    border-radius: 5px;
                    cursor: pointer;
                    color: white;
                }
                .stop-button {
                    background-color: #ff4d4d;
                    border: none;
                    padding: 10px;
                    border-radius: 5px;
                    cursor: pointer;
                    color: white;
                    margin-left: 10px;
                }
                .read-button {
                    background-color: #4caf50;
                    border: none;
                    padding: 10px;
                    border-radius: 5px;
                    cursor: pointer;
                    color: white;
                    margin-left: 10px;
                }
            </style>
        </head>
        <body>
            <div class="chat-section">
                # <div class="profile-section">
                #     <h1>Chat with Poem AI</h1>
                # </div>
                <div class="chat-messages" id="chatbox">
                    <div class="chat-bubble bot">مرحباً بك في مجلسٍ تُزهى فيه الألفاظ وتُحاكي فيه الأنغام، كيف أكون في خدمتك وأروي لك من معين الكلام الطيب؟</div>
                </div>
                <div class="chat-input-container">
                    <input type="text" id="userInput" class="chat-input" placeholder="اكتب سؤالك هنا..." />
                    <button id="sendButton" class="send-button">إرسال</button>
                    <button class="mic-button" onclick="startDictation()">🎤</button>
                    <button id="stopButton" class="stop-button" onclick="stopAudio()">إيقاف</button>
                    <button id="readButton" class="read-button" onclick="playAudio()" style="display: none;">اقرأ</button>
                </div>
            </div>

            <script>
                let audio = null;
                let currentAudioUrl = '';

                async function startDictation() {
                    if (window.hasOwnProperty('webkitSpeechRecognition')) {
                        var recognition = new webkitSpeechRecognition();
                        recognition.lang = "ar-SA"; // Arabic language
                        recognition.start();

                        recognition.onresult = function(e) {
                            document.getElementById('userInput').value = e.results[0][0].transcript;
                            recognition.stop();
                            sendMessage();
                        };

                        recognition.onerror = function(e) {
                            console.error('Speech recognition error', e);
                            recognition.stop();
                        }
                    } else {
                        alert('Speech recognition not supported in this browser.');
                    }
                }

                document.getElementById('sendButton').onclick = async function() {
                    await sendMessage();
                };

                async function sendMessage() {
                    const userInput = document.getElementById('userInput').value;
                    if (!userInput) return;

                    const chatbox = document.getElementById('chatbox');
                    chatbox.innerHTML += `<div class="chat-bubble user">أنت: ${userInput}</div>`;
                    document.getElementById('userInput').value = '';

                    const response = await fetch('/chat', {
                        method: 'POST',
                        headers: {
                            'Content-Type': 'application/json'
                        },
                        body: JSON.stringify({ question: userInput })
                    });

                    const data = await response.json();
                    const fullResponse = data.response.join("<br>");
                    chatbox.innerHTML += `<div class="chat-bubble bot">${fullResponse}</div>`;
                    chatbox.scrollTop = chatbox.scrollHeight;

                    // Store the audio URL and show the "Read" button
                    currentAudioUrl = data.audio_url;
                    document.getElementById('readButton').style.display = 'inline-block';
                };

                function playAudio() {
                    stopAudio(); // Stop any currently playing audio
                    if (currentAudioUrl) {
                        audio = new Audio(currentAudioUrl);
                        audio.play();
                        audio.onended = function() {
                            audio = null; // Clear audio when finished
                        };
                    }
                }

                function stopAudio() {
                    if (audio) {
                        audio.pause();
                        audio.currentTime = 0; // Reset audio to the start
                        audio = null; // Clear audio reference
                    }
                }
            </script>
        </body>
        </html>
    ''')

@app.route('/chat', methods=['POST'])
def chat():
    data = request.json
    question = data.get('question', '')
    response = query_model(question)

    # Generate TTS audio from the response
    audio_file = gtts_tts(" ".join(response))
    audio_url = f"/static/{os.path.basename(audio_file)}"

    return jsonify({"response": response, "audio_url": audio_url})

@app.route('/static/<path:filename>')
def static_files(filename):
    return send_from_directory('static', filename)

if __name__ == '__main__':
    # Create static directory for TTS audio files
    if not os.path.exists('static'):
        os.makedirs('static')
    app.run()
```
